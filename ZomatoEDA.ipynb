```python
ratings=final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
from collections import Counter

# Drop missing values in 'Cuisines'
cuisine_series = final['Cuisines'].dropna().str.split(', ')

# Flatten the list and count each cuisine
cuisine_counts = Counter([cuisine for sublist in cuisine_series for cuisine in sublist])

# Get the top 10 cuisines
pd.Series(cuisine_counts).sort_values(ascending=False).head(10)
```
